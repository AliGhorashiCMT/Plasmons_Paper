```julia
splits = Float64[]
x = Float64[]
for id in 1:51
    charge=(round(-1+(id-1)/50, digits=3))
    N = 1 .+ charge
    println(charge)
    filename = "Lithium/Li-4-4-$charge.eigenvals"
    
    filename_nup = "Lithium/Li-4-4-$charge.n_up"
    
    filename_ndn = "Lithium/Li-4-4-$charge.n_dn"
        
    n_up = np.fromfile(filename_nup)
    n_dn = np.fromfile(filename_ndn)
    
    diff = abs(sum((n_up.^(1/3) - n_dn.^(1/3))))

    push!(x, N^(1))
    println(filename)
    data = JJDFTX.bandstruct_properties(filename, 3, 19, spin=2);
    push!(splits, abs(mean(data[2][3]) - mean(data[1][3])))

end
```
