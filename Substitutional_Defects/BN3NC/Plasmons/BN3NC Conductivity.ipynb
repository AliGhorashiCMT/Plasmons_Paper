```julia
phonon11_dir = "../phonon11/"
phonon22_dir = "../phonon22/";
phonon33_dir = "../phonon33/";
```

```julia
Hwannier, cell_map = hwannier(phonon22_dir*"wannier"), np.loadtxt(phonon22_dir*"wannier.map.txt");
lattice_vectors = loadlattice(phonon22_dir*"wannier.out");
Pwannier = pwannier(phonon22_dir*"wannier");
```

```julia
Energies, DOS = density_of_states(Hwannier, cellmap, Val(2), histogram_width=100,
    mesh=20, monte_carlo=true, num_blocks=500, degeneracy=2)
```

```julia
Es, Cs = find_chemical_potential(Energies, DOS)
μ = Energies[argmin(abs.(Cs .- 1))]
println("Chemical potential: $(μ)")
println("gμ: $(DOS[argmin(abs.(Energies .- μ))])")
```

```julia
# Drude Conductivity
es, cs = drude_conductivity(lattice_vectors, Hwannier, cell_map, Pwannier, mesh=20, num_blocks=1000,
    degeneracy=2, histogram_width=20)
```

```julia
(cs[argmin(abs.(es .- μ)), 2, 2]+cs[argmin(abs.(es .- μ)), 1, 1])/2
```
