```julia
for i in 0:9
    for j in 0:9
        x_diff = i/10
        y_diff = j/10
        open("BN22BC_$(i)_$(j).ionpos", "w") do io 
            for line in readlines("../BN22BC_Buckled/BN22BC.ionpos")[2:end-1]
                 _, ion, x, y, z, _= split(line)
                x, y, z = round.(parse.(Float64, [x, y, z]), digits=4)
                write(io, "ion $ion $(x) $(y) $(z) 0\n")
                write(io, "ion $ion $(x+x_diff) $(y+y_diff) $(z+1/2) 0 \n")
            end
        end
    end
end
```
