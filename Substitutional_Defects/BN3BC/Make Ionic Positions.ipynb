```julia
lattice_vectors = [1 -1/2; 0 sqrt(3)/2]
supercell_vectors = [3/2 0; -sqrt(3)/2 sqrt(3)]
```

```julia
# The conversion from a vector in lattice coordinates to superlattice coordinates
inv(supercell_vectors)*lattice_vectors
```

```julia
# From the above, it is clear that (n, m) in lattice coordinates is mapped to (2/3n-1/3m, 1/3n+1/3m)
# This maps on to 3 possibilities (0, 0), (2/3, 1/3), (1/3, 2/3)
```

```julia
#Position of Nitrogen atom
round.(inv(supercell_vectors)*lattice_vectors*[1/3, 2/3], digits=3)
```
