```julia
Strains=[]
for (i, line) in enumerate(readlines("hBN-convergence.out"))
    if contains(line, "Stress tensor in Cartesian coordinates")
        x_strain = abs(parse(Float64, split(readlines("hBN-convergence.out")[i+1])[2]))
        y_strain = abs(parse(Float64, split(readlines("hBN-convergence.out")[i+2])[3]))
        z_strain = abs(parse(Float64, split(readlines("hBN-convergence.out")[i+3])[4]))
        push!(Strains, x_strain+y_strain+z_strain)
    end
end
```

```julia
plot(Strains)
```
