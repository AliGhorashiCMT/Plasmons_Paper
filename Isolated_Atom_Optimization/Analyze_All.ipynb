```julia
using PyPlot, PyCall, Peaks
np=pyimport("numpy")
```

```julia
li_dir="./Lithium"
na_dir="./Sodium"
cs_dir="./Cesium"
rb_dir="./Rubidium"
k_dir="./Potassium";
```

```julia
Li_Energies=[]
for i in 1:100
    push!(Li_Energies, parse(Float64, split(readlines(`listEnergy $li_dir/Lithium-$i.out`)[1])[1]))
end
```

```julia
Cs_Energies=[]
for i in 1:100
    push!(Cs_Energies, parse(Float64, split(readlines(`listEnergy $cs_dir/Cesium-$i.out`)[1])[1]))
end
```

```julia
K_Energies=[]
for i in 1:100
    push!(K_Energies, parse(Float64, split(readlines(`listEnergy $k_dir/Potassium-$i.out`)[1])[1]))
end
```

```julia
Na_Energies=[]
for i in 1:100
    push!(Na_Energies, parse(Float64, split(readlines(`listEnergy $na_dir/Sodium-$i.out`)[1])[1]))
end
```

```julia
Rb_Energies=[]
for i in 1:100
    push!(Rb_Energies, parse(Float64, split(readlines(`listEnergy $rb_dir/Rubidium-$i.out`)[1])[1]))
end
```

```julia
dists = round.(0.5 .+ collect(1:100) ./ 100*3, digits=3);
```

```julia
plot(dists,log.(Na_Energies .- minimum(Na_Energies) .+ 0.001), label="Sodium", linewidth=5)
plot(dists,log.(Li_Energies .- minimum(Li_Energies) .+ 0.001), label="Lithium", linewidth=5)
plot(dists,log.(K_Energies .- minimum(K_Energies) .+ 0.001), label="Potassium", linewidth=5)
plot(dists, log.(Rb_Energies .- minimum(Rb_Energies) .+ 0.001), label="Rubidium", linewidth=5)
plot(dists, log.(Cs_Energies .- minimum(Cs_Energies) .+ 0.001), label="Cesium", linewidth=5)

title("Equilibrium Triangular Lattice Size")
xlabel("Multiplicity of hBN Lattice")
#xlim(1, 2)
legend()
```
