```julia
using PyCall, PyPlot
```

```julia
na_dir="./Sodium"
li_dir="./Lithium"
k_dir="./Potassium"
```

```julia
Na2 = zeros(101);
Na3 = zeros(101);
Na4 = zeros(101);
```

```julia
Li2 = zeros(101);
Li3 = zeros(101);
Li4 = zeros(101);
```

```julia
K2 = zeros(101);
K3 = zeros(101);
K4 = zeros(101);
```

```julia
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Li2[i] = parse(Float64, split(readlines(`listEnergy $li_dir/Li-2-$mag.out`)[1])[1])
    Li3[i] = parse(Float64, split(readlines(`listEnergy $li_dir/Li-3-$mag.out`)[1])[1])
    Li4[i] = parse(Float64, split(readlines(`listEnergy $li_dir/Li-4-$mag.out`)[1])[1])
end
```

```julia
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    K2[i] = parse(Float64, split(readlines(`listEnergy $k_dir/K-2-$mag.out`)[1])[1])
    K3[i] = parse(Float64, split(readlines(`listEnergy $k_dir/K-3-$mag.out`)[1])[1])
    K4[i] = parse(Float64, split(readlines(`listEnergy $k_dir/K-4-$mag.out`)[1])[1])
end
```

```julia
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Na2[i] = parse(Float64, split(readlines(`listEnergy $na_dir/Na-2-$mag.out`)[1])[1])
end
```

```julia
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Na3[i] = parse(Float64, split(readlines(`listEnergy $na_dir/Na-3-$mag.out`)[1])[1])
end
```

```julia
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Na4[i] = parse(Float64, split(readlines(`listEnergy $na_dir/Na-4-$mag.out`)[1])[1])
end
```

```julia
figure(figsize=(15, 10))
subplot(2, 3, 1)
plot(Na2, linewidth=5)
plot(Na3, linewidth=5)
plot(Na4, linewidth=5)
subplot(2, 3, 2)
plot(Li2, linewidth=5)
plot(Li3, linewidth=5)
plot(Li4, linewidth=5)
subplot(2, 3, 3)
plot(K2, linewidth=5)
plot(K3, linewidth=5)
plot(K4, linewidth=5)
```
