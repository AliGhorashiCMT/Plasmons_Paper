```julia
using PyCall, PyPlot, Statistics
eV = 27.2
```

```julia
na_dir="./Sodium";
li_dir="./Lithium";
k_dir="./Potassium";
rb_dir="./Rubidium";
cs_dir="./Cesium";
```

```julia
mags = -1:0.02:1
```

```julia
Na2 = zeros(101);
Na3 = zeros(101);
Na4 = zeros(101);
```

```julia
Li2 = zeros(101);
Li3 = zeros(101);
Li4 = zeros(101);
```

```julia
K2 = zeros(101);
K3 = zeros(101);
K4 = zeros(101);
```

```julia
Rb2 = zeros(101);
Rb3 = zeros(101);
Rb4 = zeros(101);
```

```julia
Cs2 = zeros(101);
Cs3 = zeros(101);
Cs4 = zeros(101);
```

```julia
#Lithium
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Li2[i] = parse(Float64, split(readlines(`listEnergy $li_dir/Li-2-$mag.out`)[1])[1])*eV
    Li3[i] = parse(Float64, split(readlines(`listEnergy $li_dir/Li-3-$mag.out`)[1])[1])*eV
    Li4[i] = parse(Float64, split(readlines(`listEnergy $li_dir/Li-4-$mag.out`)[1])[1])*eV
end
```

```julia
#Potassium
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    K2[i] = parse(Float64, split(readlines(`listEnergy $k_dir/K-2-$mag.out`)[1])[1])*eV
    K3[i] = parse(Float64, split(readlines(`listEnergy $k_dir/K-3-$mag.out`)[1])[1])*eV
    K4[i] = parse(Float64, split(readlines(`listEnergy $k_dir/K-4-$mag.out`)[1])[1])*eV
end
```

```julia
#Sodium
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Na2[i] = parse(Float64, split(readlines(`listEnergy $na_dir/Na-2-$mag.out`)[1])[1])*eV
    Na3[i] = parse(Float64, split(readlines(`listEnergy $na_dir/Na-3-$mag.out`)[1])[1])*eV
    Na4[i] = parse(Float64, split(readlines(`listEnergy $na_dir/Na-4-$mag.out`)[1])[1])*eV
end
```

```julia
#Rubidium
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Rb2[i] = parse(Float64, split(readlines(`listEnergy $rb_dir/Rb-2-$mag.out`)[1])[1])*eV
    Rb3[i] = parse(Float64, split(readlines(`listEnergy $rb_dir/Rb-3-$mag.out`)[1])[1])*eV
    Rb4[i] = parse(Float64, split(readlines(`listEnergy $rb_dir/Rb-4-$mag.out`)[1])[1])*eV
end
```

```julia
#Cesium
for i in 1:101
    mag = round(-1+(i-1)*2/100, digits=3)
    Cs2[i] = parse(Float64, split(readlines(`listEnergy $cs_dir/Cs-2-$mag.out`)[1])[1])*eV
    Cs3[i] = parse(Float64, split(readlines(`listEnergy $cs_dir/Cs-3-$mag.out`)[1])[1])*eV
    Cs4[i] = parse(Float64, split(readlines(`listEnergy $cs_dir/Cs-4-$mag.out`)[1])[1])*eV
end
```

```julia
figure(figsize=(15, 10))
subplot(2, 3, 1)
plot(mags, Na2 .- mean(Na2), linewidth=5, color="red")
plot(mags, Na3 .- mean(Na3), linewidth=5, color="black")
plot(mags, Na4 .- mean(Na4), linewidth=5, color="blue")
xlabel("Magnetization (Bohr)")

ylabel("")
title("Sodium")
subplot(2, 3, 2)
plot(mags, Li2 .- mean(Li2), linewidth=5, color="red")
plot(mags, Li3 .- mean(Li3), linewidth=5, color="black")
plot(mags, Li4 .- mean(Li4), linewidth=5, color="blue")
xlabel("Magnetization (Bohr)")

title("Lithium")
subplot(2, 3, 3)
plot(mags, K2 .- mean(K2), linewidth=5, color="red")
plot(mags, K3 .- mean(K3), linewidth=5, color="black")
plot(mags, K4 .- mean(K4), linewidth=5, color="blue")
xlabel("Magnetization (Bohr)")

title("Potassium")
subplot(2, 3, 4)
plot(mags, Rb2 .- mean(Rb2), linewidth=5, color="red")
plot(mags, Rb3 .- mean(Rb3), linewidth=5, color="black")
plot(mags, Rb4 .- mean(Rb4), linewidth=5, color="blue")
xlabel("Magnetization (Bohr)")
title("Rubidium")

subplot(2, 3, 5)
plot(mags, Cs2 .- mean(Cs2), linewidth=5, color="red")
plot(mags, Cs3 .- mean(Cs3), linewidth=5, color="black")
plot(mags, Cs4 .- mean(Cs4), linewidth=5, color="blue")
xlabel("Magnetization (Bohr)")
title("Cesium")
```
